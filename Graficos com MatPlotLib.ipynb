```python
temperatura_por_dia_da_semana = dados.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo','Segunda','Terça','Quarta','Quinta','Sexta','Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias] # Altera o indice?
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize = (10,5))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana)) # criando um indice do tamanho dos dias da semana

eixo.bar(indice, temperatura_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize = 20)
eixo.set_xlabel('Dias', fontsize = 16)
eixo.set_ylabel('Temperatura média', fontsize = 16)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias) # mudando o nome dos ticks do eixo x
plt.show()
```

```python
fig = plt.figure(figsize = (10,5))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana)) # criando um indice do tamanho dos dias da semana
cores = ['black','r','g','b','yellow','orange','magenta']  # adicionando cores

eixo.bar(indice, temperatura_por_dia_da_semana, color = cores)
eixo.set_title('Temperatura por dia da semana', fontsize = 20)
eixo.set_xlabel('Dias', fontsize = 16)
eixo.set_ylabel('Temperatura média', fontsize = 16)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias) # mudando o nome dos ticks do eixo x
plt.show()
```

```python
fig = plt.figure(figsize = (16,8))
eixo = fig.add_axes([0,0,1,1])

# scatter é gráfico de dispersão
eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Gráfico de Dispersão', fontsize = 20, pad = 10)
eixo.set_xlabel('Comprimento', fontsize = 16)
eixo.set_ylabel('Largura', fontsize = 16)
eixo.tick_params(labelsize = 15)  # aumenta o tamanho da fonte dos parametros dos eixos x e y

plt.show()
```

```python
fig = plt.figure(figsize = (16,8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa':'r','Iris-versicolor':'b','Iris-virginica':'g'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'],tmp['largura_sépala'], color = cores[especie])

# scatter é gráfico de dispersão
eixo.set_title('Gráfico de Dispersão', fontsize = 20, pad = 10)
eixo.set_xlabel('Comprimento', fontsize = 16)
eixo.set_ylabel('Largura', fontsize = 16)
eixo.tick_params(labelsize = 15)  # aumenta o tamanho da fonte dos parametros dos eixos x e y
eixo.legend(cores, fontsize = 16)

plt.show()
```

```python
fig = plt.figure(figsize = (16,8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa':'r','Iris-versicolor':'b','Iris-virginica':'g'}
marcadores = {'Iris-setosa':'x','Iris-versicolor':'o','Iris-virginica':'v'}  # criando os marcadores

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'],tmp['largura_sépala'], color = cores[especie], marker = marcadores[especie])

# scatter é gráfico de dispersão
eixo.set_title('Gráfico de Dispersão', fontsize = 20, pad = 10)
eixo.set_xlabel('Comprimento', fontsize = 16)
eixo.set_ylabel('Largura', fontsize = 16)
eixo.tick_params(labelsize = 15)  # aumenta o tamanho da fonte dos parametros dos eixos x e y
eixo.legend(cores, fontsize = 16)

plt.show()
```
