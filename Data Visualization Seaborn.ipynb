## Teste de HipÃ³tese

**H<sup>null</sup>**
>**A distribuiÃ§Ã£o da taxa da gorjeta Ã© a mesma nos dois grupos**

**H<sup>alt</sup>**
>**A distribuiÃ§Ã£o da taxa da gorjeta nÃ£o Ã© a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
dados = dados.rename(columns = {'porcentagem %':'porcentagem'})
sobremesa = dados.query('sobremesa == "Sim"')['porcentagem']
```

```python
sem_sobremesa = dados.query('sobremesa == "NÃ£o"')['porcentagem']
```

```python
r = ranksums(sobremesa,sem_sobremesa)
r
```

```python
print('O valor do pvalue Ã© {}, entÃ£o deve-se utilizar a hipotese nula.'.format(r.pvalue))
```

### A distribuiÃ§Ã£o da taxa de gorjetas dos dois grupos Ã© a mesma

## Teste de Hipotese

**H<sup>null</sup>**
>**A distribuiÃ§Ã£o do valor da conta Ã© igual no sÃ¡bado e no domingo**

**H<sup>alt</sup>**
>**A distribuiÃ§Ã£o do valor da conta nÃ£o Ã© igual no sÃ¡bado e no domingo**

```python
conta_domingo = dados.query('dia == "Domingo"')['total_conta']
conta_domingo
```

```python
conta_sabado = dados.query('dia == "SÃ¡bado"')['total_conta']
conta_sabado
```

```python
r2 = ranksums(conta_domingo, conta_sabado)
r2
```

```python
print('O valor do pvalue Ã© {}'.format(r2.pvalue))
```

### A hipÃ³tese vÃ¡lida Ã© a nula, entÃ£o o valor da conta Ã© igual no sÃ¡bado e no domingo
