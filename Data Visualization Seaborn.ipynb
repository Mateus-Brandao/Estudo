## Teste de Hipótese

**H<sup>null</sup>**
>**A distribuição da taxa da gorjeta é a mesma nos dois grupos**

**H<sup>alt</sup>**
>**A distribuição da taxa da gorjeta não é a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
dados = dados.rename(columns = {'porcentagem %':'porcentagem'})
sobremesa = dados.query('sobremesa == "Sim"')['porcentagem']
```

```python
sem_sobremesa = dados.query('sobremesa == "Não"')['porcentagem']
```

```python
r = ranksums(sobremesa,sem_sobremesa)
r
```

```python
print('O valor do pvalue é {}, então deve-se utilizar a hipotese nula.'.format(r.pvalue))
```

### A distribuição da taxa de gorjetas dos dois grupos é a mesma

## Teste de Hipotese

**H<sup>null</sup>**
>**A distribuição do valor da conta é igual no sábado e no domingo**

**H<sup>alt</sup>**
>**A distribuição do valor da conta não é igual no sábado e no domingo**

```python
conta_domingo = dados.query('dia == "Domingo"')['total_conta']
conta_domingo
```

```python
conta_sabado = dados.query('dia == "Sábado"')['total_conta']
conta_sabado
```

```python
r2 = ranksums(conta_domingo, conta_sabado)
r2
```

```python
print('O valor do pvalue é {}'.format(r2.pvalue))
```

### A hipótese válida é a nula, então o valor da conta é igual no sábado e no domingo
