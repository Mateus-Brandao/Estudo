```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3
```

```python
# criando um banco de dados
Conexao = sqlite3.connect('Banco de Dados')
```

```python
# apontar para o banco de dados
Cursor = Conexao.cursor()
```

```python
# criando uma tabela
Cursor.execute(
    "CREATE TABLE Minha_Tabela (Data text, Nome text, Idade real)"
)

# fazer um comit 
Conexao.commit()
```

```python
# inserindo valores
Cursor.execute('INSERT INTO Minha_Tabela VALUES("03/06/2022", "Mateus","26")')
```

```python
# importando numero aleatorios
import random
```

```python
# loop
for Loop in range(10):
  # gerando um numero aleatorio
  Numero = random.randint(10,20)
  # inserindo informação na tabela
  Cursor.execute(f'INSERT INTO Minha_Tabela VALUES ("22/03/2018", "Banguela", {Numero})')

```

```python
# Query de consulta
Consulta = Cursor.execute('SELECT * FROM Minha_Tabela').fetchall()

print(Consulta)
```

```python
# Query de consulta
Consulta = Cursor.execute('SELECT Nome, Idade FROM Minha_Tabela').fetchall()
for linha in Consulta:
  print(linha)
```

```python
# Query de consulta
Consulta = Cursor.execute('SELECT Data FROM Minha_Tabela').fetchall()

print(Consulta)
```

```python
# Query com where
Consulta = Cursor.execute('''
SELECT * FROM Minha_Tabela
WHERE Idade = '20'
''').fetchall()

print(Consulta)
```

```python
# Query com where
Consulta = Cursor.execute('''
SELECT * FROM Minha_Tabela
WHERE Idade >= '20'
''').fetchall()

for linha in Consulta:
 print(linha)
```

```python
# Query com between
Consulta = Cursor.execute('''
SELECT * FROM Minha_Tabela
WHERE Idade BETWEEN 18 AND 20
''').fetchall()

for linha in Consulta:
 print(linha)
```

```python
# inserindo valores nulos
Cursor.execute('INSERT INTO Minha_Tabela VALUES(null, "Mateus","26")')
Cursor.execute('INSERT INTO Minha_Tabela VALUES("03/06/2022", null,null)')
```

```python
# Query com isnull
Consulta = Cursor.execute('''
SELECT * FROM Minha_Tabela
WHERE Nome IS NULL
''').fetchall()

for linha in Consulta:
 print(linha)
```

# FUNÇÃO UPDATE 

```python
# Query geral
Consulta = Cursor.execute('''
SELECT * FROM Minha_Tabela
''').fetchall()

for linha in Consulta:
 print(linha)
```

```python
# Verificando valores não nulos
Consulta = Cursor.execute('''
UPDATE Minha_Tabela
SET Nome = 'Preenchido'
WHERE Nome IS NULL
''').fetchall()

Consulta = Cursor.execute('''
SELECT * FROM Minha_Tabela
''').fetchall()

for linha in Consulta:
 print(linha)
```

```python
Consulta = Cursor.execute('''
UPDATE Minha_Tabela
SET Idade = 30
WHERE Idade IS NULL
''').fetchall()

Consulta = Cursor.execute('''
SELECT * FROM Minha_Tabela
''').fetchall()

for linha in Consulta:
 print(linha)
```
