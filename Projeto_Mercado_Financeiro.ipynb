## Séries Temporais

```python
# criando uma nova base com a data com index para facilitar algumas análises
dados = db.set_index(['Data'])
dados.head()
```

```python
# criando uma análise gráfica do valor de fechamento durante o ano de 2021
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,8))
plt.title('Análise das ações - Fechamento', fontsize = 16)
plt.xlabel('Periodo da cotação', fontsize = 14)
plt.ylabel('Valor da cotação em R$', fontsize = 14)

plt.plot(dados.index, dados['Fechamento']);
```

### É possível notar que o valor das ações caiu consideravelmente durante o ano de 2021

```python
# criando uma média movel de 5 dias
media_movel = dados['Fechamento'].rolling(5).mean()
media_movel
```

```python
# criando um média tendência de 30 dias
media_tendencia = dados['Fechamento'].rolling(30).mean()
media_tendencia
```

```python
# criando um gráfico com os valores de fechamento, média movel e média tendência durante o ano de 2021
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,8))
plt.title('Análise das ações - Fechamento', fontsize = 16)
plt.xlabel('Periodo da cotação', fontsize = 14)
plt.ylabel('Valor da cotação em R$', fontsize = 14)

plt.plot(dados.index, dados['Fechamento'])
plt.plot(dados.index, media_movel)
plt.plot(dados.index, media_tendencia);
```

```python
# criando um boxplot do fechamento
plt.figure(figsize=(12,6))
sns.boxplot(data = db, x = 'Fechamento')
```

```python
# criando a coluna mês na base de dados
db['Mês'] = db['Data'].dt.month
db.head()
```

```python
# analisando o box plot do fechamento por mês
plt.figure(figsize=(12,6))
sns.boxplot(data = db, x = 'Mês', y = 'Fechamento')
```

```python
# descrevendo os dados do fechamento por mês
db.groupby(by = ['Mês']).describe()['Fechamento']
```

```python
 # criando um grafico tipo Candlestick da bovespa 
 grafico = go.Figure(data = [go.Candlestick(x = db.index, open = dados['Abertura'], high = dados['Maior'],
                                            low = dados['Menor'], close = dados['Fechamento'])])
 
grafico.update_layout(xaxis_rangeslider_visible = False)

grafico.show()
```
